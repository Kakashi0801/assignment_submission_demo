```python
# For plotting the data
import matplotlib.pyplot as plt
import seaborn as sns
# For data manipultaion
import pandas as pd
# For scientific or mathematical computations
import numpy as np
```

```python
np.random.rand(1000, 2)
```

```python
x1 =  np.random.rand(1000, 2)
y1 =  3 *x1* np.random.randn(1000, 2)
```

```python
plt.scatter(x1, y1)
plt.title("Scatter plot between $X$ and $Y$")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
y1 = np.array([50, 100, 150, 200, 250])
plt.plot(y1)
plt.title("Line plot")
plt.xlabel("$X$ axis")
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([20,40,60,80,100,120,140,160,180,200,220,240,260,280])
plt.ylabel("$Y$ axis")
plt.show()
```
